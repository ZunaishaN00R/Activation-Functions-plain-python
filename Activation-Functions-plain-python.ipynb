1.	**Create Tanh, ReLU and Leaky ReLU activation functions by their formulas.**

**Hyperbolic Tangent**

```python
import math
```

```python
def hyperbolic(z):
  x=math.exp(z) - math.exp(-z)
  y=math.exp(z) + math.exp(-z)
  tanh=x/y
  return tanh
```

```python
hyperbolic(-56)
```

**ReLU**

```python
import math
```

```python
def relu(x):
  return max(0,x)
```

```python
relu(-100)
```

```python
relu(8)
```

**Leaky ReLU**

```python
def leaky_relu(x):
  return max(0.1*x,x)
```

```python
leaky_relu(-100)
```
